## Solution

```python
class MarkovDecisionProcess:
    def __init__(
        self, states: list, actions: set, gamma: numbers.Real,
        transition_probability_matrices: numpy.ndarray, reward_matrices: numpy.ndarray
    ) -> None:
        self.states = states
        self.actions = actions
        self.gamma = gamma
        self.transition_probability_matrices = transition_probability_matrices
        self.reward_matrices = reward_matrices

    def evaluate_policy(self, policy: dict[str: numbers.Real], threshold: numbers.Real = 1e-2):
        '''Runs iterative policy evaluation algorithm.'''

        # 1. Initialise value functions at 0
        value_function_vec = numpy.zeros(len(self.states))

        # 2. Iterate until convergence
        delta = -numpy.inf
        iter_num = 1
        while delta < 0 or delta > threshold:
            print(f'Iteration #{iter_num}')
            # 3. Evaluate value function for each state
            for state_num, state_name in enumerate(self.states):
                cur_v = bellman_equation(  # TODO: error?
                    state_num=state_num, 
                    policy=policy,
                    transition_probability_matrices=transition_probability_matrices,
                    reward_matrices=reward_matrices,
                    value_function_vec=value_function_vec,
                    gamma=self.gamma
                )
                value_function_vec[state_num] = cur_v # TODO: error?
                # print(
                #     pandas.Series(value_function_vec, index=self.states)
                # )
                cur_iter_diff = abs(cur_v - value_function_vec[state_num])
                delta = max([delta, cur_iter_diff])
            print(f'Delta: {delta}')
            iter_num += 1

        return value_function_vec
    

def bellman_equation(
        state_num: int, 
        policy: dict[str: numbers.Real],
        transition_probability_matrices: dict[str: numpy.ndarray],
        reward_matrices: dict[str: numpy.ndarray],
        value_function_vec: numpy.ndarray,
        gamma: numbers.Real
    ):
    v = 0
    for policy_name, policy_prob in policy.items():

        cur_tpm = transition_probability_matrices[policy_name]
        cur_reward_matrix = reward_matrices[policy_name]

        cur_tpm_vector = cur_tpm[state_num, :].reshape(-1, 1) 
        cur_reward_vector = cur_reward_matrix[state_num, :].reshape(-1, 1)
        cur_vfv = value_function_vec.reshape(-1, 1)
        cur_action_value =\
            policy_prob * ( cur_tpm_vector.T @ (cur_reward_vector + gamma * cur_vfv) )[0, 0]
        v += cur_action_value

    return v
```

```python
mdp = MarkovDecisionProcess(
    states=states, actions=actions, gamma=gamma,
    transition_probability_matrices=transition_probability_matrices,
    reward_matrices=reward_matrices
)
```
